```python
import pandas as pd

# Load the Titanic dataset
titanic_data = pd.read_csv('titanic_survival_data.csv')

# Selecting relevant columns: "Pclass", "Sex", "Age", "Fare", and "Survived"
X = titanic_data[['Pclass', 'Sex', 'Age', 'Fare']]
y = titanic_data['Survived']
```

```python
# Replacing missing values in "Age" and "Fare" with the mean of the respective column
X['Age'].fillna(X['Age'].mean(), inplace=True)
X['Fare'].fillna(X['Fare'].mean(), inplace=True)
```

```python
# Convert the categorical variables "Pclass" and "Sex" into dummy variables
X_dummies = pd.get_dummies(X, columns=['Pclass', 'Sex'], drop_first=True)
```
